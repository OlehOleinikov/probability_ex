```python
import itertools
import numpy as np

all_combinations = list(itertools.product([0, 1], repeat=6))
filtered_combinations = [comb for comb in all_combinations if comb.count(1) == 2]
for comb in filtered_combinations:
    print(comb)
```

```python
len(filtered_combinations)
```

```python
def prob_func(aim_event, other_events):
    """Класична ймовірність = сприятливі ділити на суму сприятливих та несприятливих"""
    return aim_event/(aim_event + other_events)

events = {0: 'normal', 1: 'special'}

prob_list = []  # тут накопичуються ймовірності кожної окремої комбінації

for iter_num, comb_set in enumerate(filtered_combinations):  # ітерації по наборам комбінацій
    print(f'Combination {iter_num+1}: {comb_set}')

    # Кількість кульок на початок перевірки комбінації (потім будемо виймати)
    normal_items = 8  #
    special_items = 4 #
    print(f'Initial set: normal - {normal_items}, special - {special_items}')

    res_list = []  # тут накопичуються ймовірності кожного елемента комбінації (з урахуванням витягування кульок)

    for step in comb_set: # ітерація по кожному елементу комбінації
        if step == 0:
            prob = prob_func(normal_items, special_items)
            print(f'\tCurrent item: "{events[step]}"  '
                  f'items available |{normal_items} normal| |{special_items} special| '
                  f'(prob = {prob:.3f})')
            res_list.append(prob)
            normal_items -= 1  # виймаємо відповідну кульку
        else:
            prob = prob_func(special_items, normal_items)
            print(f'\tCurrent item: "{events[step]}" '
                  f'items available |{normal_items} normal| |{special_items} special| '
                  f'(prob = {prob:.3f})')
            res_list.append(prob)
            special_items -= 1 # виймаємо відповідну кульку

    print(f'\t\tProb array  : {res_list}')
    comb_prob = np.prod(np.array(res_list)) # множення всіх елементів (послідовні вірогідності у поточній комбінації)
    print(f'\t\tProb product: {comb_prob}\n')
    prob_list.append(comb_prob)

print('\nTotal results:')
for pos, pr in enumerate(prob_list):
    print(f'Comb#{pos+1:2d}: {pr:.5f}')
print('Sum=', sum(prob_list))


```
